Con el comando open abrimos nuestro archivo de tal manera que las filas se almacenan
como elementos de una lista, la lista “leer

```python
posicion=[]
constante = leer[0].find('KWBC')
print(constante)
for i in range(len(leer)):  
  if leer[i].find('KWBC') == constante:
    posicion.append(i)
posicion.append(len(leer)) 
print(posicion)
```

```python
tamañofilasf450=[]
for i in range(len(posicion)-1):
  tamañofilasf450.append(len(leer[posicion[i+1]-5]))

print(tamañofilasf450)
```

eliminar repetidas

```python
tamfilas=[]
for i in tamañofilasf450:
  if i not in tamfilas:
    tamfilas.append(i)
print(tamfilas)
```

ordenar

```python
tamañodefilas=sorted(tamfilas)
print(tamañodefilas)
```

Ya con las posiciones mencionadas anteriormente, nos ubicamos 2 filas antes de cada
una de estas, que es dónde se encuentran las filas “F450”, para poder medir su tamaño
y luego formar los casos correspondientes. El resultado nos arroja 3 tamaños: 50, 59 y
68

```python
listadireccion=[]
listavelocidad=[]

for j in range(len(posicion)-1):
  for i in range(posicion[j],posicion[j+1]):
    if leer[i].find('F450') == 0:
      if len(leer[i])==50:
        listadireccion.append(leer[i][5:8]+' '+leer[i][14:17]+' '+leer[i][23:26]+' '+leer[i][32:35]+' '+leer[i][41:44]+' ')
        listavelocidad.append(leer[i][8:10]+' '+leer[i][17:19]+' '+leer[i][26:28]+' '+leer[i][35:37]+' '+leer[i][44:46]+' ')
      elif len(leer[i])==59:
        listadireccion.append(leer[i][5:8]+' '+leer[i][14:17]+' '+leer[i][23:26]+' '+leer[i][32:35]+' '+leer[i][41:44]+' '+leer[i][50:53]+' ')
        listavelocidad.append(leer[i][8:10]+' '+leer[i][17:19]+' '+leer[i][26:28]+' '+leer[i][35:37]+' '+leer[i][44:46]+' '+leer[i][53:55]+' ')
      elif len(leer[i])==68:
        listadireccion.append(leer[i][5:8]+' '+leer[i][14:17]+' '+leer[i][23:26]+' '+leer[i][32:35]+' '+leer[i][41:44]+' '+leer[i][50:53]+' '+leer[i][59:62]+' ')
        listavelocidad.append(leer[i][8:10]+' '+leer[i][17:19]+' '+leer[i][26:28]+' '+leer[i][35:37]+' '+leer[i][44:46]+' '+leer[i][53:55]+' '+leer[i][62:64]+' ')

for i in listadireccion:
  print(i)

print('\n*********** \n')
for i in listavelocidad:
  print(i)
```
